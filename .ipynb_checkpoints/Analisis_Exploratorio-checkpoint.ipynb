Para ver de forma más detallada esta distribución anómala (antes de la segregación por especie) podemos tomar una de las gráficas. En este caso, la medida de la aleta:

```python
sns.histplot(
    data = process_penguin_df,
    x= 'flipper_length_mm',
    kde = True,
    bins = 18
)

plt.axvline(
    x = process_penguin_df.flipper_length_mm.mean(),
    color = 'red',
    linestyle = 'dashed',
    linewidth = 2
)
plt.axvline(
    x = process_penguin_df.flipper_length_mm.median(),
    color = 'green',
    linestyle = 'dashed',
    linewidth = 2
)

plt.axvline(
    x = process_penguin_df.flipper_length_mm.quantile(0.25),
    color = 'blue',
    linestyle = 'dashed',
    linewidth = 2
)

plt.axvline(
    x = process_penguin_df.flipper_length_mm.quantile(0.75),
    color = 'black',
    linestyle = 'dashed',
    linewidth = 2
)

plt.title('Fliper Length', loc = 'left')
plt.text(process_penguin_df.flipper_length_mm.mean(), 53, 'Mean', fontsize=12, color = 'red')
plt.text(process_penguin_df.flipper_length_mm.median(), 50, 'Median', fontsize=10, color = 'green')
plt.text(process_penguin_df.flipper_length_mm.quantile(0.25), 53, 'Q(25%)', fontsize=12, color = 'blue')
plt.text(process_penguin_df.flipper_length_mm.quantile(0.75), 53, 'Q(75%)', fontsize=12, color = 'black')
```

En este ejemplo se hace más claro el sesgo de la distribución a la izquierda, pero al analizar la gráfica nuevamente, agregando la segmentación por especie, vemos tres distribuciones de datos que se acercan un poco más a la distribución normal.

**Aplicando funciones de probabilidad**

**¿Cómo se distribuye la variable peso? ¿Es posible ajustar una distribución teórica a la distribución empírica?**

Teniendo en cuenta el aprendizaje anterior, vamos a realizar este ajuste de la variable peso tomando la especie Adelie como referencia.

```python
stats = process_penguin_df.query("species == 'Adelie'").body_mass_g.describe()
stats
```

```python
xs = np.linspace(stats['min'], stats['max'])
ys = scipy.stats.norm(stats['mean'], stats['std']).cdf(xs)

plt.plot(xs, ys, color = 'black', linestyle = 'dashed')

#La línea punteada muestra el comportamiento de las probabilidades si los datos siguieran una distribución normal
#Graficamos la función de probabilidad acumulada

empiricaldist.Cdf.from_seq(
    process_penguin_df.query("species == 'Adelie'").body_mass_g,
    normalize = True
).plot() #Línea azul
```

```python
xs = np.linspace(stats['min'], stats['max'])
ys = scipy.stats.norm(stats['mean'], stats['std']).pdf(xs)

plt.plot(xs, ys, color = 'black', linestyle = 'dashed')

#La línea punteada muestra el comportamiento de las probabilidades si los datos siguieran una distribución normal
#Graficamos la función de probabilidad acumulada

sns.kdeplot(
    data = process_penguin_df.query("species == 'Adelie'"),
    x = 'body_mass_g'
)
```

A partir de estos dos contrastes entre la distribución teórica y la real, 
observamos que es posible acercarnos a la distribución empírica del peso de los pingüinos Adelies, lo que nos permite simular datos.

```python
sex_numeric = process_penguin_df.sex.replace(['male', 'female'], [0, 1])
```

```python
number_samples = 100
sample_size = 35

samples_df = pd.DataFrame()

np.random.seed(42)

for i in range(1, number_samples + 1):
    sex_numeric_sample = sex_numeric.sample(sample_size, replace = True).to_numpy()
    sample_name = f"sample{i}"
    samples_df[sample_name] = sex_numeric_sample

male_population_mean = samples_df.mean().mean()

print(f"Porcentaje estimado de pinguinos machos en la población es: {male_population_mean * 100:.4f}%")
```

```python
sample_means_binomial = pd.DataFrame(samples_df.mean(), columns =['sample_mean'])
sns.kdeplot(data = sample_means_binomial)
plt.axvline(x = sex_numeric.mean(), color = 'r', linestyle = '--')
```

El gráfico de puntos nos permite analizar la dispersión de una variable respecto a otra, para este primer caso tomaremos la longitud de los picos vs la profundidad de los mismos.

En la gráfica anterior no parece haber un patrón en el contraste de las dos variables mencionadas. Incluso, de haberlo, parecería una correlación negativa, es decir, ante el aumento de la longitud del pico se presenta una disminución de su profundidad, ¿esto tiene sentido?

**Estableciendo relaciones: Análisis de Regresión Simple**

Por medio del Análisis de Regresión Simple podemos estimar la relación de aquellas variables que previamente hemos identificado como variables correlacionadas. No obstante, para hacer hincapié en la importancia de analizar bien el comportamiento de los datos para no caer en la [Paradoja de Simpson](https://es.wikipedia.org/wiki/Paradoja_de_Simpson), haremos primero el análisis de regresión simple entre la longitud y la profundidad del pico sin segregarlo por especie.

```python
sns.scatterplot(
    data = process_penguin_df,
    x = 'bill_length_mm',
    y = 'bill_depth_mm'
)

res_bill_penguins = scipy.stats.linregress(
    x = process_penguin_df.bill_length_mm,
    y = process_penguin_df.bill_depth_mm
)

print(res_bill_penguins)

fx_1 = np.array([process_penguin_df.bill_length_mm.min(), process_penguin_df.bill_length_mm.max()])
fy_1 = res_bill_penguins.intercept + res_bill_penguins.slope * fx_1

plt.plot(fx_1, fy_1)
```

Tal como se oberva, la regresión lineal que se plantea es una línea con pendiente negativa, con lo que se podría decir que a medida que aumenta la longitud del pico, disminuye su profundidad, pero esto se desmiente en el siguiente gráfico, en el cual se calcula las regresiones lineales por cada especie con las mismas dos variables.

```python
#Graficando la regresión directamente con Seaborn

sns.lmplot(
    data = process_penguin_df,
    x = 'bill_length_mm',
    y = 'bill_depth_mm',
    height = 10,
    hue = 'species'
)
```

Tal como se esperaba, hay una correlación positiva entre estas variables teniendo en cuenta las características de cada especie.

En el caso de las variables **Peso** y **Longitud de las alas** vemos que la correlación es evidente y al graficar una regresión lineal simple obtenemos una línea con pendiente positiva, independiente del tipo de especie de los pingüinos, esto nos servirá para ir llegando al objetivo de este análisis: Definir modelos predictivos.

```python
#Regresión lineal entre el peso y el tamaño de las alas

sns.scatterplot(
    data = process_penguin_df,
    x = 'body_mass_g',
    y = 'flipper_length_mm'
)


reg_body_flip = scipy.stats.linregress(
    x = process_penguin_df.body_mass_g,
    y= process_penguin_df.flipper_length_mm)

print(reg_body_flip)


eje_x_regress_bf = np.array([process_penguin_df.body_mass_g.min(), process_penguin_df.body_mass_g.max()])
eje_y_regress_bf = reg_body_flip.intercept + reg_body_flip.slope * eje_x_regress_bf

plt.plot(eje_x_regress_bf, eje_y_regress_bf)
```

## **Modelos predictivos**

Teniendo en cuenta el análisis realizado y la exploración del comportamiento de las variables, procederemos a plantear modelos predictivos y definir el que nos dé un porcentaje de presición más alto.

Para empezar, trataremos de crear un modelo que prediga el peso de los pingüinos a partir de las demás observaciones.

**Creando un modelo de Regresión Lineal**

```python
#Variable dependiente: Body_mass_g, variable independiente: fliper_length_mm

(
    smf.ols(
        formula = 'body_mass_g ~ flipper_length_mm',
        data = process_penguin_df
    )
    .fit()
    .summary()
)
```

**Análisis de regresión múltiple**

Partiendo del supuesto de que no tengo forma de conocer el peso de los próximos pingüinos en nuestro estudio,
¿Cómo podría capturar este dato?

**Creando modelos**

*Modelo 1*

```python
#Variable dependiente: Body_mass_g, variable independiente: bill_length_mm

model_1 = (
    smf.ols(
    formula = 'body_mass_g ~ bill_length_mm',
    data = process_penguin_df
    ).
    fit()
)

model_1.summary()
```

Notas para interpretación: El 86.79 del coeficiente de la longitud del pingüino nos indica que por cada cambio en dicha medida, es decir, en la longitud del pico, esperaríamos un aumento de 86.79 gramos en el peso.

```python
#Variable dependiente: Body_mass_g, variables independientes: bill_length_mm y bill_depth_mm

model_2 = (
    smf.ols(
    formula = 'body_mass_g ~ bill_length_mm + bill_depth_mm',
    data = process_penguin_df
    ).
    fit()
)

model_2.summary()
```

```python
#Variable dependiente: Body_mass_g, variables independientes: bill_length_mm, bill_depth_mm y flipper_length_mm

model_3 = (
    smf.ols(
    formula = 'body_mass_g ~ bill_length_mm + bill_depth_mm +  flipper_length_mm',
    data = process_penguin_df
    ).
    fit()
)

model_3.summary()
```

```python
#Variable dependiente: Body_mass_g, variables independientes: bill_length_mm, bill_depth_mm, flipper_length_mm y sex

model_4 = (
    smf.ols(
    formula = 'body_mass_g ~ bill_length_mm + bill_depth_mm +  flipper_length_mm + C(sex)',
    data = process_penguin_df
    ).
    fit()
)

model_4.summary()
```

```python
#Variable dependiente: Body_mass_g, variables independientes: flipper_length_mm y sex
model_5 = (
    smf.ols(
    formula = 'body_mass_g ~ flipper_length_mm + C(sex)',
    data = process_penguin_df
    ).
    fit()
)

model_5.summary()
```

**Visualizando resultados de los Modelos**

Una vez generados los modelos, procedemos a crear un DataFrame para observar los resultados de la predicción y analizarlos gráficamente con el fin de elegir nuestro modelo.

Creación de tabla de resultados.

```python
model_results = pd.DataFrame(
    dict(
        actual_value = process_penguin_df.body_mass_g,
        prediction_model_1 = model_1.predict(),
        prediction_model_2 = model_2.predict(),
        prediction_model_3 = model_3.predict(),
        prediction_model_4 = model_4.predict(),
        prediction_model_5 = model_5.predict(),
        species = process_penguin_df.species,
        sex = process_penguin_df.sex
    )
)

model_results.head()
```

**Graficando los resultados con la Función de Distribución Acumulada**

```python
sns.ecdfplot(
    data = model_results
)
```

Si bien varios modelos tienen un comportamiento similar a los valores reales (actual_value), nos basaremos en el valor del R-cuadrado para definir el indicado, pues un R-cuadrado cercano a 1 nos indica que la calidad del modelo es alta y este puede explicar una gran parte de la variabilidad de de la variable dependiente.

Los modelos con mejor R-cuadrado con el Modelo 4 (0.82) y el Modelo 5 (0.80). Si bien el modelo 4 tiene mayor R-cuadrado, seleccionaré el modelo 5 dado que también parece confiable y depende de menos variables, lo que puede ser más útil en un caso de información faltante.

```python
#Graficando solo los datos reales vs el modelo 5

sns.ecdfplot(
    data = model_results.select_columns(['actual_value', 'prediction_model_5'])
)
```

**Analizando los resultados desde la Probabilidad de densidad**

```python
sns.kdeplot(
    data = model_results.select_columns(['actual_value', 'prediction_model_5']),
    #cumulative = True
)
```

**Análisis de Regresión Logística**

¿Podemos crear un modelo que nos ayude a definir si un pingüino es macho o hembra? Para este caso utilizaremos un modelo de regresión logística.

```python
#Variable dependiente: Sexo - Variables independientes: flipper_length_mm, bill_length_mm, bill_depth_mm y la Isla de Origen

model_1_sex= smf.logit(
    formula = 'numeric_sex ~ flipper_length_mm + bill_length_mm + bill_depth_mm + C(island)',
    data = process_penguin_df
).fit()

model_1_sex.summary()
```

```python
model_2_sex= smf.logit(
    formula = 'numeric_sex ~ flipper_length_mm + bill_length_mm',
    data = process_penguin_df
).fit()

model_2_sex.summary()
```

**¿Podemos definir un modelo que nos ayude a identificar si un modelo pertenece a una determinada especie?**

Dado que los modelos de Regresión logística son binarios, no puedo definir un modelo que me clasifique un pingüino entre las tres diferentes opciones, lo que puedo hacer es definir un modelo que prediga si un pingüino pertenece o no a una especie determinada. Por lo tanto, podemos hacer un modelo que prediga si un pingüino pertenece o no a una especie, en este caso tomaremos la especie Adelie.

```python
process_penguin_df.species.unique()
```

```python
#Para este caso, tomaré la especie Adelie

process_penguin_df = process_penguin_df.assign(
    is_adelie = lambda df: df.species.replace(['Adelie', 'Gentoo', 'Chinstrap'], [1, 0, 0])
)
```

```python
model_is_adelie = smf.logit(
    formula = 'is_adelie ~ flipper_length_mm + C(sex)',
    data = process_penguin_df
).fit()

model_is_adelie.params
```

```python
is_adelie_df_predictions = pd.DataFrame(
    dict(
        actual_adelie = process_penguin_df.is_adelie,
        predicted_values = model_is_adelie.predict().round()
    )
)

is_adelie_df_predictions
```

```python
(
    is_adelie_df_predictions
    .value_counts(['actual_adelie', 'predicted_values'])
    .reset_index(name = 'count')
)

#El count nos da el resultado de las predicciones de acuerdo al cruce mostrado en la fila
```

Analizando la fiabilidad del modelo con Sklearn

```python
print(
    sklearn.metrics.confusion_matrix(
        is_adelie_df_predictions.actual_adelie,
        is_adelie_df_predictions.predicted_values
    )
)

sklearn.metrics.accuracy_score(
    is_adelie_df_predictions.actual_adelie,
    is_adelie_df_predictions.predicted_values
)
```

    Falso Negativo (1,0): 17 (Error Tipo II) - el modelo los clasificó como de otra especie diferente y son Adelie
